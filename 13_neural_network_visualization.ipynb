```python
# import module
import graphviz
import os
```

```python
# instantiating object
dot = graphviz.Digraph(comment='A Round Graph')
# Adding nodes
dot.node('A', 'Fruits')
dot.node('B', 'Apple')
dot.node('C', 'Orange')
dot.node('D', 'Grapes')
# Adding edges
dot.edges(['AB', 'AC', 'AD'])
dot.edge('B', 'C', constraint='false')
dot.edge('C', 'D', constraint='false')

print(dot.source)
# saving source code
dot.format = 'png'
dot.render('Graph', view=True)
```
