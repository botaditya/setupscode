```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Loading Iris dataset
df = pd.read_csv("./datasets/Iris.csv")
df.head(10)
df.shape

#Calculate Correlation
print('Correlation')
df.corr(method='pearson')
```

```python
#Plot Heatmap
# adds the title
plt.title('Correlation using heatmap')
sns.heatmap(df.corr(method='pearson').drop(
    ['Id'], axis=1).drop(['Id'], axis=0),
    annot=True)
plt.show()

#Plot Colorbar
plt.matshow(df.corr(method='pearson'))
plt.colorbar()

# Overview of Relationships using Pairplot
sns.pairplot(df.drop(['Id'], axis=1),
             hue='Species', height=2)
```
